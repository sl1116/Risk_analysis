```python
from sklearn.ensemble import RandomForestClassifier

# Build random forest model and calculate feature importances
rf = RandomForestClassifier(n_estimators=100, random_state=42)
rf.fit(X, y)
importances = rf.feature_importances_

# Calculate Information Value (IV) for each feature
ivs = []
for i in range(len(importances)):
    iv = sum([(x / sum(importances)) * (importances[x] - importances[i])**2 for x in range(len(importances))])
    ivs.append(iv)

# Filter out features with IV less than 0.02
selected_features = list(filter(lambda x: ivs[X.columns.get_loc(x)] >= 0.02, X.columns))
new_df = df[selected_features]

# Save new dataset to file
new_df.to_csv('UCI_Credit_Default_Filtered.csv', index=False)
```

```python
def calc_woe(df, col, target):
    total_good = df[target].value_counts()[0]
    total_bad = df[target].value_counts()[1]
    bins = np.linspace(df[col].min(), df[col].max(), 11)
    groups = df.groupby(pd.cut(df[col], bins=bins))
    woe_dict = {}
    for name, group in groups:
        good = group[target].value_counts().get(0, 0)
        bad = group[target].value_counts().get(1, 0)
        if good != 0 and bad != 0:
            woe = np.log((good / total_good) / (bad / total_bad))
        else:
            woe = 0
        woe_dict[name] = woe
    return woe_dict

woe_dict = calc_woe(df, 'LIMIT_BAL', 'default.payment.next.month')
df['LIMIT_BAL_WOE'] = pd.cut(df['LIMIT_BAL'], bins=np.linspace(df['LIMIT_BAL'].min(), df['LIMIT_BAL'].max(), 11)).map(woe_dict)
print(df[['LIMIT_BAL', 'LIMIT_BAL_WOE']].head())
```

```python
# Create a bar chart of WOE values for each bin of a variable
def plot_woe(df, col, target):
    woe_dict = calc_woe(df, col, target)
    fig, ax = plt.subplots(figsize=(8, 6))
    x_labels = [str(interval) for interval in woe_dict.keys()]
    x_ticks = np.arange(len(x_labels))
    y_values = list(woe_dict.values())
    ax.bar(x_ticks, y_values, align='center', color='b')
    ax.set_xticks(x_ticks)
    ax.set_xticklabels(x_labels, rotation=45, ha='right')
    ax.set_title(f'WOE values for {col} bins')
    ax.set_xlabel(f'{col} bins')
    ax.set_ylabel('WOE value')
    plt.show()

# Plot WOE values for each variable in the dataset
for col in df.columns[:-1]:
    plot_woe(df, col, 'default.payment.next.month')
```

```python
# Set the values of odds and corresponding credit scores for scaling
odds_100 = 50
score_100 = 100
odds_120 = 100
score_120 = 120

# Calculate the offset and factor based on the two data points
offset = score_100 - np.log(odds_100)
factor = (score_120 - np.log(odds_120) - offset) / np.log(odds_120)

# Define a scaling function using the offset and factor
def scale_credit_score(x):
    return offset + factor * np.log(x)

# Apply the scaling function to LIMIT_BAL column in new_df
new_df['LIMIT_BAL_scaled'] = new_df['LIMIT_BAL'].apply(scale_credit_score)

# Print the first few rows of the scaled dataset
print(new_df.head())
```
